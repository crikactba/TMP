#### 1. Carregue os dados da tarefa anterior

````python
from pandas_datareader import data as pdr

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
````

````python
companies = ['AMZN','GOOG','TSLA'] 
tickers = yf.Tickers(companies)
end_date = datetime.now().strftime('%Y-%m-%d')
tickers_hist = tickers.history(start='2022-01-01',end=end_date)
tickers_hist
````

````python
tickers = tickers_hist.stack(level=1).rename_axis(['Date', 'Ticker']).reset_index(level=1)
tickers
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
yf_close             = tickers_hist['Close'].dropna()
yf_close_90d         = yf_close.rolling(90).mean()
yf_close_90d         = yf_close_90d.stack().reset_index()
yf_close_90d.columns = ['Data','Ativo','Valor']
yf_close_90d
````

````python
yf_close_desvio         = yf_close.rolling(90).std()
yf_close_desvio         = yf_close_desvio.stack().reset_index()
yf_close_desvio.columns = ['Data','Ativo','Valor']
yf_close_desvio
````

````python
# grafico media movel 90 dias
fig = px.line(yf_close_90d, x='Data', y="Valor", color='Ativo', template='plotly_white')
fig.show()
````

````python
# grafico desvio padrão
fig = px.line(yf_close_desvio, x='Data', y="Valor", color='Ativo', template='plotly_white')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
indicadores = ['Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['AMZN'] * len(indicadores)
colunas = list(zip(indicadores, ativo))
amazon = tickers_hist.loc[tickers_hist.index[-60]:,colunas]
amazon.columns = indicadores
````

````python
from plotly import graph_objects
from plotly.graph_objs import Layout
````

````python
graph_data = {
    'x': amazon.index,
    'open': amazon.Open,
    'close': amazon.Close,
    'high': amazon.High,
    'low': amazon.Low,
    'type': 'candlestick',
    'name': 'AMZN',
    'showlegend': True
}
graph_layout = Layout(
    paper_bgcolor='rgba(255,255,255,255)',
    plot_bgcolor='rgba(255,255,255,255)'
)
````

````python
graph = graph_objects.Figure(
    data=[graph_data],
    layout_title='AMAZON',
    layout=graph_layout
)
graph.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
graph.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
graph
````
